# Exploratory Analysis

## Import libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Read data

Use pandas to read data. Use 'id' as index.

```python
cancer_data = pd.read_csv('../data/Cancer_Data.csv', sep=',', index_col=0).iloc[:,:-1]
```

View data.

```python
cancer_data
```

## Check for missing values

You can see that there are no missing values in each column.

```python
cancer_data.isnull().any()
```

## View data distribution

Set image style.

```python
sns.set_style('ticks')
sns.set(font_scale=0.8)
```

Use the frequency histogram to see how the data is distributed. It is found that the data in all numerical columns are normally distributed.

```python
num_data = cancer_data.iloc[:, 1:]
num_data.reset_index(inplace=True, drop=True)
n_rows = 5
n_cols = 6
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10))
for i in range(n_rows):
    for j in range(n_cols):
        col_index = i * n_cols + j
        if col_index < 30:
            sns.histplot(num_data.iloc[:, col_index], ax=axes[i, j], kde=True)
plt.tight_layout()
plt.savefig('../figure/distribution.svg', dpi=600, format='svg')
plt.show()
```
